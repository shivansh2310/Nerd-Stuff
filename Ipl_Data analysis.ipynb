```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ipl_new=pd.read_csv('/content/IPL Matches 2008-2020.csv')
```

```python
ipl_new['toss_winner'].value_counts()
```

```python
batting_first=ipl_new[ipl_new['result'] !='wickets']
```

```python
batting_first.head()
```

```python
plt.figure(figsize=(5,7))
plt.hist(batting_first['result_margin'])
plt.title("Distribution of runs")
plt.xlabel("Runs")
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]), color=["blue", "yellow", "red"]) 
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%') 
plt.show()
```

```python
batting_second=ipl_new[ipl_new['result'] !='runs']
```

```python
batting_second.head()
```

```python
plt.figure(figsize=(5,7))
plt.hist(batting_second['result_margin'], bins=30)
plt.show()
```

```python
batting_second['winner'].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]), color=["Purple", "blue", "red"]) 
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%') 
plt.show()
```
